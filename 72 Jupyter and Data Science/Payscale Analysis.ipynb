```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
df = pd.read_csv('payscale_data.csv')
for column in ['Early Career Pay', 'Mid-Career Pay']:
    df[column] = [float(data[1:].replace(",", "")) for data in df[column]]
df['% High Meaning'] = [int(data[:-1]) if data != "-" else None for data in df['% High Meaning'] ]

df.head()
```

```python
df = df.sort_values('% High Meaning', ascending=False)
df.head()
```

```python
music = df.loc[df['Major'].str.contains('Music')]
pay_meaning = music['Mid-Career Pay'] / music['% High Meaning']
music.insert(2, "Pay Meaning", pay_meaning)
```
